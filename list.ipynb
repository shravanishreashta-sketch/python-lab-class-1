```python
#tuple
print("tuple example")
cordinates=(10.5,4.5,5.4)
print("tuple:",cordinates)
print("first cordinate:",cordinates[0])
print("last  cordinate:",cordinates[-1])#gives last element for -1
```

```python
#tuple unpacking
x,y,z= cordinates
print(f"unpacked:x={x},y={y},z={z}")
```
