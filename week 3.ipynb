```python
#program to find largest number

num = int(input("Enter a number: "))
if num >= 0:
    print("Non-negative number")
else:
    print("Negative number")

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Largest:", a)
elif b > c:
    print("Largest:", b)
else:
    print("Largest:", c)


```

```python
# Sum of N numbers
n = int(input("Enter value of n: "))
sum = 0
for i in range(1, n + 1):
    sum += i
print("Sum =", sum)
```

```python
#MULTIPLICATION TABLE
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# For loop example
for i in range(1, 6):
    print(i)
```

```python
# Factorial
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial =", fact)
```

```python
# While loop example
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
# Count digits
num = int(input("Enter a number: "))
count = 0
while num != 0:
    num = num // 10
    count += 1
print("Number of digits:", count)
```

```python
# Program to check even or odd
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even Number")
```
