```python
#Basic callback
def process (data,callback):
    result=data*2
    callback(result)
def showoutput(value):
      print("processed:",value)
process(5,showoutput)
```

```python
#function returning multiple values
def stats(numbers):
    return min (numbers), max (numbers)/len (numbers)
print(stats([2,4,6,8]))
```

```python
#function with parameters
def add(a,b):
    return a + b
print(add(3,5))
```

```python
def power(base,exp=2):
    return base** exp
print(power(4))
print(power(4,3))
```
